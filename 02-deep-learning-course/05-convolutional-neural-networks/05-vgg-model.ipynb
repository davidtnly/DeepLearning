```python
import numpy as np
from keras.applications import vgg16
from keras.preprocessing import image
```

```python
model = vgg16.VGG16(weights='imagenet')
```

```python
img = image.load_img('Course_Images/spoon.jpeg', target_size=(224, 224))
img
```

```python
# Convert to a numpy array
img_array = image.img_to_array(img)
img_array.shape
```

```python
# Expand the dimension
img_array = np.expand_dims(img_array, axis=0)
img_array.shape
```

```python
# Preprocessing
img_array = vgg16.preprocess_input(img_array)
img_array
```

```python
# Predict
prediction = model.predict(img_array)
prediction
```

```python
# Top 5 predictions
vgg16.decode_predictions(prediction, top=5)
```

### Multiple Images

```python
from keras.models import model_from_json
from keras.preprocessing import image
from pathlib import Path
from PIL import Image
import numpy as np
import joblib

########################################
#### Multiple Images
########################################

# Training folder
images_folder = 'Other_Images/'

# Create path to folders with training data
img_path = Path(images_folder)

# Create empty list
images = []

# Loop through images
for img in img_path.glob('*.png'): # return a possibly-empty list of path names that match pathname "*.png"
    
    Image.open(img)
    print(img)
    
    # Load the image from disk
    img = image.load_img(img, target_size=(224, 224))
    
    # Convert the image to a numpy array
    image_array = image.img_to_array(img) # keras helper function to convert image into an array
    
    # Expand dimensions (np)
    image_array = np.expand_dims(image_array, axis=0)    
    
    # Preprocessing (vgg16)
    image_array = vgg16.preprocess_input(image_array)
    
    # Predict (model)
    predictions = model.predict(image_array)
    
    # Top 5 predictions
    vgg16.decode_predictions(predictions, top=5)
```
