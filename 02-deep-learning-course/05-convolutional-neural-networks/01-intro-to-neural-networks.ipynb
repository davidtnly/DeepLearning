## Perceptron

A perceptron is a basic artificial neural network. 

Perceptron or an artificial neuron computes a weighted sum of the inputs. So we have weight W one multiplied by input X one with W two multiplied by input X two and so on, and then we add a bias b. The weighted sum is then fed into an __activation function__ to determine whether it is triggered or not. 

This is basically a __binary classifier__ because what we are seeing is that if it exceeds a certain threshold, return a one, otherwise return a zero. There needs to be a way for our network to learn the values of b and W and quantify any error. We'll be looking at that in our next video on gradient descent.

### Loss
When we looked at the perceptron, or artificial neuron, we said that it was made up of weighted sum of inputs. The inputs are x1, x2, all the way up to xn, and the weights are w1, w2, all the way to wn.

With a classification problem, for a given input, we know what the actual output is. And when we pass the same input into our model, we then have a predicted output. 

We call the difference between these two ouputs the __loss__, and to improve our model's prediction capability, we want to minimize this loss. We can change our predictions by changing our weights within the model. 


## Gradient Descent

### Learning a Neuron's Weight / Parameters

Now one of the things we need to do is to learn a neuron's weights and parameters. These are the weights and the biases. We can use a loss function to quantify the loss. We take the gradient of this loss function to determine where we can minimize the loss. 

__Gradient descent__ is __an optimizing algorithm__ which is used to iterate through different combinations of weights to find the best combinations of weights which minimizes this loss or error. 

```python
Image.open('Images/02-gradient-descent.png')
```

So if you have some way of quantifying the error your network makes on training data, which is the differential of loss function, then we can take the gradient of this loss function, with respect to the weights, and update our weights slowly in the direction of the gradient. 

### Finding The Minimum of The Loss
This allows us to find the minimum of the loss function, and then we get better predictions. 

So there are two things to look out for. 
1. Which direction to go to get the minimum of the __loss__ function, and the __learning rate__, which tells us how big the step should be to get there. 

### Descent Side
So if you're on the left of the curve, we know that the gradient or slope of the line is negative, and to achieve the minimum we want, we will have to move the ball further down the curve. And if, on the other hand, the gradient is positive, we know that we will have to move the ball to the left of the curve. 

When the ball is at the bottom of the slope, we have reached the lowest position, and __minimized the loss function__. 

## Applying Gradient Descent

### Challenge
The gradient is not differentionable at certain points. What we mean by that: the gradient, or slope, is the change in y divided by the change in x, and if you wanted to find the slope at a point where the output goes from zero to one, you're going from zero to one, but not moving along in the x-axis. 

So you are dividing a number by zero, and we say this isn't defined mathematically. You cannot differentiate the step function at certain points, so we need to find a way around this. 

### Solution

A work-around is to use the __logistic sigmoid function__, and like the previous network, it will give a value of zero or one, where the cutoff mark could be the halfway point. 

The sigmoid function looks like this, and you can see that it is like the step function, but looks smoothened out, and more importantly, we can differentiate it.

## XOR Challenge

Visualizing just X1 and X2, we should be able to separate these two outputs with a single line. It's very clear that you can separate the two classes of data here. But no matter what you try and do, you can't draw a straight line to separate the output. 

A single artificial neuron (perceptron) can only learn linearly separable functions and XOR can't be separted in q 

Researchers soon found that using a binary classifier such as a single artificial neuron, there was no linear function that could be applied that would separate the blue from the black. 

Now this was a big problem because clearly the black and the blue can be easily separated or classified. This is known as the __XOR problem__ and researchers assumed that there was no value in investing time in AI research if it couldn't solve a simple problem like this. 

This was responsible for what is termed the cold and dark winter of artificial intelligence research from 1970 to 1985.

### What is an XOR function and why is it significant? 

The XOR, or Exclusive OR, is a Boolean operator, like and or, and these are the foundations of computer logic. With XOR, you can do parity checks and swaps without using a temporary variable. It is clear we need to introduce some nonlinear functionality.

### Multilayer Perceptrons

We saw some of the limitations of using a single perceptron, as the output is only a linear combination of the inputs, and how __we need to introduce non-linearity__ into our network by using non-linear activation functions.

### Key Question
One of the key questions then is how do we extend from going from a single neuron to a Neural Network. We will define a __Multilayer Perceptron__, as being one where the perceptrons are divided up into multiple layers. 
- The output of a perceptron will be the input to another perceptron. You will see that this is a fully connected layer, and so each layer processes all of the outputs from the previous layer. 


- Each of the perceptrons in the diagram behave like the single perceptron. 


- The input is a linear combination of the sum the products of the weight times the input. The input to this perceptron is a linear combination of all of the outputs multiplied by weight, which I have indicated by Wij, of perceptrons of the previous layer. 


- The output from this perceptron is connected to all of the perceptrons in the next layer. 


- We introduced the non-linearity by insuring that the activation function we used, the function that determines if the neuron triggers or not, is non-linear. 


So, we have an input layer and an output layer, and if you have any layers in between, as shown in this diagram, this is known as a hidden layer. Here, we only have one hidden layer, but in reality there could be multiple hidden layers. A deep Neural Network is then a Neural Network with one or more hidden layers.
