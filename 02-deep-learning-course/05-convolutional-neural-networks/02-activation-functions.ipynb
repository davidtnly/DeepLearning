# Activation Functions

The activation function determines if a neuron fires or not. 

```python
from PIL import Image
Image.open('Images/07-activation-function.png')
```

### Nonlinearity

To ensure that there is some nonlinearity in our network, we need to make sure that these activation functions are nonlinear. 

### Step Function

We could use the step function as our activation function, which provides an output of zero or one. If the output is above a certain threshold, then neuron is fired, and we have a one. If the value of the output is less than the threshold, then it is not fired, and we have a zero. 

```python
Image.open('Images/08-step-function.png')
```

### What if our output needs to be multiple classes and not just two

We can use a function that provides a number of discrete outputs of classes. For example, answers could range from zero to nine, instead of being zero and one, as it is in the step function. We don't want to use a linear function as an activation function, as we will end up with the XOR problem. 

Additionally, multiple layers with linear activation functions can be replaced by a single linear function. With the __sigmoid function__, the output is between zero and one. This means that this is good for binary classifications as your output is one, and zero, or true and false. 

```python
Image.open('Images/09-sigmoid.png')
```

### Problem with Sigmoid

The problem is what is known as __vanishing gradient__. Near the boundaries, the network doesn't learn quickly. This is because the gradient or slope is almost zero at both ends.

### TanH (Hyperbolic Tangent)

 Tanh or hyperbolic tangent is another possible nonlinear activation function. Sigmoid has an output between zero and one, __TanH is defined in terms of exponents__, and has an output between minus one and one. As the outputs are not one and zero, or true and false, it doesn't lend itself as well to binary classification problems as sigmoid. 
 
Additionally, a tanh function can be made up of sigmoids, meaning that because sigmoids have problems with the vanishing gradient problem, this will apply to tanh too. 

### ReLU (Rectified Linear Unit)

The ReLU or rectified linear unit is defined as outputting a zero for any value of x that is less than zero. For any value of x equal to or greater than zero, the function returns an x. 

So for example, if your input x was minus two, then the ReLU would return zero. If your input x was plus two instead, then the ReLU function would return plus two. 

```python
Image.open('Images/10-relu.png')
```

### Which one to use?
At first glance, you might think that the ReLU has the same problems as linear functions. However, ReLU is nonlinear, and combinations of ReLU are nonlinear. As you can imagine, it'll be less computationally expensive than say a sigmoid function. ReLU is one of the more well-known and used activation functions. 

So how do you know which one to use? You'll need to experiment and see which one provides the best performance for your particular model. But in general, you will find that ReLU provides a good performance in many instances, and is a good starting point.

## Backpropagation & Hyperparameters

What backpropagation tries to achieve is quite simple. We have the input to the network, and our neural network provides some output. We then want to compare that to the expected output or the actual output. We compute the loss function and take a gradient of the loss function. We can then compute it for the weights of the output neuron directly.

```python
Image.open('Images/11-forward-prop.png')
```

### Where does the backwards part come from?

The backwards part of the name is because the calculation of the gradient proceeds backwards throughout the network, with the gradient of the final layer of weights being calculated first and the gradient of the first layer of weights being calculated last. 

Partial computations of the gradient from one layer are reused in the computation of the gradient for the previous layer. This backward flow of the error information allows for efficient computation of the gradient at each layer.
